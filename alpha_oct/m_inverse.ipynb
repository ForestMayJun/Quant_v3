## 测试上午和下午的动量和反转情况

```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
tqdm.pandas
```

```python
close23 = pd.read_hdf('/mnt/datadisk2/aglv/aglv/lab_aglv/data2023.h5')['Close']
close23
```

```python
eg = close23.unstack()
eg
```

```python
min30 = eg.groupby(level='Date').apply(lambda x: x.iloc[:, 30] / x.iloc[:, 1] - 1)
min30
```

```python
min30.index.names = ['a', 'Date', 'InstrumentID']
idnew = min30.index.droplevel('a')
min30.index = idnew
min30 = min30.unstack()
min30
```

```python
min30_updown = min30 / min30.abs()
min30_updown.sum(axis=0)
```

```python
morn_p = eg.loc[:, :'11:30']
morn_p
```

```python
morn_ret = morn_p.groupby(level='Date').apply(lambda x: x.iloc[:, -1] / x.iloc[:, 1] -1)
morn_ret
```

```python
# morn_ret.index.names = ['a', 'Date', 'InstrumentID']
# eg = morn_ret.index.droplevel('a')
morn_ret.index = eg
morn_ret =  morn_ret.unstack()
morn_ret
```

```python
afno_p = eg.loc[:, '13:00':]
afno_p
```

```python
afno_ret = afno_p.groupby(level='Date').apply(lambda x: x.iloc[:, -1] / x.iloc[:, 0] -1)
afno_ret.index.names = ['a', 'Date', 'InsrumentID']
afno_ret.index = afno_ret.index.droplevel('a')
afno_ret = afno_ret.unstack()
afno_ret
```

```python
eg = afno_ret.corrwith(morn_ret)
eg[eg > 0]
```
