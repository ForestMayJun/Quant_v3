```python
from DataDaily import DataDaily
datadaily = DataDaily()
```

```python
close = datadaily.adjclose
ret = close.pct_change()
ret20 = ret.rolling(20).mean()
ret20
```

```python
buy_l_order = datadaily.buy_value_large_order.loc[20210101:20231231].T
buy_l_order
```

```python
buy_l_order_demean = buy_l_order / buy_l_order.rolling(20).mean()
buy_l_order_demean
```

```python
sell_l_order = datadaily.sell_value_large_order.loc[20210101:20231231].T
sell_l_order_demean = sell_l_order / sell_l_order.rolling(20).mean()
# del sell_l_order
# sell_l_order_demean
```

```python
buy_xl_order = datadaily.buy_value_exlarge_order.loc[20210101:20231231].T
buy_xl_order
```

```python
sell_xl_order = datadaily.sell_value_exlarge_order.loc[20210101:20231231].T
sell_xl_order
```

```python
buy_mean_devide_std = buy_l_order_demean.rolling(10).max() / buy_l_order_demean.rolling(10).std()
buy_mean_devide_std
```

```python
b_and_s_squaresum = (np.power(buy_l_order_demean, 2) + np.power(sell_l_order_demean, 2)) / np.power((buy_l_order_demean + sell_l_order_demean), 2)
b_and_s_squaresum
```

```python
b_and_s_squaresum.rolling(10).mean().to_csv('/mnt/datadisk2/aglv/aglv/aglv_factor/Oct_factor/b_ane_s_square_sum_r1mean.csv')
```

```python
buy_mean_devide_std.to_csv('/mnt/datadisk2/aglv/aglv/aglv_factor/Oct_factor/buy_max_devide_std_r20.csv')
```

```python
ret20_b_and_s_squaresum = ret20 * b_and_s_squaresum
ret20_b_and_s_squaresum.rolling(10).mean().to_csv('/mnt/datadisk2/aglv/aglv/aglv_factor/Oct_factor/ret20_b_ane_s_square_sum_r1mean.csv')
```

```python
b_s_order = pd.concat([buy_l_order.stack(), sell_l_order.stack(), buy_xl_order.stack(), sell_xl_order.stack()], axis=1)
b_s_order
```
