```python
def get_corr(df:pd.DataFrame):
    res = pd.DataFrame(index=df.index.get_level_values('Date').unique(), columns=df.columns)
    for day in tqdm(df.index.get_level_values('Date').unique().to_list()):
        data_oneday = df.xs(day, level='Date')
        data_oneday = (data_oneday - data_oneday.mean()) / (data_oneday.std() + 1e-10)
        res.loc[day] = data_oneday.corrwith(data_oneday.shift(-1).fillna(0))
        
    return res
```
