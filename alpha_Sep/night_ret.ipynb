```python

def cVaR_series(series, c_level=0.9, is_low=True):
    '''
    CVaR:分布的尾部积分, Parameters:
    c_level:p值,默认0.9
    is_low:返回的是否是左侧CVaR,=False时返回右侧,默认True
    '''

    if is_low:
        left_VaR = series.quantile(1- c_level)
        X_low = pd.Series(series[series < left_VaR])
        low_mean = X_low.mean() if len(X_low) > 0 else 0
        return low_mean
    else:
        right_VaR = series.quantile(c_level)
        X_high = pd.Series(series[series > right_VaR])
        high_mean = X_high.mean() if len(X_high) > 0 else 0
        return -high_mean
```

```python
tqdm.pandas()
from concurrent.futures import ThreadPoolExecutor

# 多线程 apply 函数
def parallel_rolling_apply(df, window_size, func, num_threads=4):
    rolling_windows = df.rolling(window=window_size)  # 创建 rolling 对象
    
    # 将每个窗口任务分发到不同的线程中
    with ThreadPoolExecutor(max_workers=num_threads) as executor:
        results = list(executor.map(func, [window for window in rolling_windows]))
    
    # 将结果转换为 pandas Series 并返回
    return pd.Series(results, index=df.index)

nightret_conti = night_ret_r3mean.progress_apply(lambda x: parallel_rolling_apply(x, 20, cVaR_series))
nightret_conti
```
