```python
from forintern.DataDaily import DataDaily
datadaily = DataDaily() 
```

```python
close = datadaily.adjclose
open = datadaily.adjopen
vol = datadaily.volume
```

```python
vol
```

```python
vol_mean20 = vol / vol.rolling(20).mean()
vol_mean20
```

```python
close_pct = close.pct_change()
for i in range(10):
    print(close_pct.shift(i).corrwith(vol_mean20).mean())
```
