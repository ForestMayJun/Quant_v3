```python
def turn_std(vol:pd.DataFrame):
    def _turn_std(vol_daily:pd.DataFrame):
        vol_daily_percent = vol_daily / vol_daily.sum()
        vol_std = vol_daily_percent.std()

        return vol_std
    
    return vol.groupby(level='Date').progress_apply(_turn_std)
```

```python
def turn_std_v4(vol:pd.DataFrame, agg_num=10):
    def _assist(vol_daily:pd.DataFrame):
        vol_daily /= vol_daily.sum()
        vol_daily.index = list(range(len(vol_daily)))
        vol_daily_agg = vol_daily.groupby(vol_daily.index // agg_num).sum()

        return vol_daily_agg.std()
    
    return vol.groupby(level='Date').progress_apply(_assist)
```

```python
def turn_std_v5(vol:pd.DataFrame, agg_num=10):
    def _assist(vol_daily:pd.DataFrame):
        vol_daily /= vol_daily.sum()
        vol_daily.index = list(range(len(vol_daily)))
        vol_daily_agg = vol_daily.groupby(vol_daily.index // agg_num).sum()

        return vol_daily_agg.diff().abs().sum()
    
    return vol.groupby(level='Date').progress_apply(_assist)
```

```python
def factor_group(factor, group_number=10):
    '''按照因子值分组,grp1最小'''
    # fg = factor.apply(
    #     lambda x: pd.qcut(x.rank(method='first'), q=group_number, labels=False) + 1, 
    #     axis=1)
    def categorize_with_nan(x, group_number):
        if x.isna().any():
            # 保留 NaN 的逻辑
            return pd.Series(pd.qcut(x.rank(method='first').fillna(-1), q=group_number, labels=False, duplicates='drop') + 1).replace(-1, np.nan)
        else:
            return pd.qcut(x.rank(method='first'), q=group_number, labels=False) + 1
    
    fg = factor.apply(lambda x: categorize_with_nan(x, group_number), axis=1)
    
    def _factor_group(factor_daily:pd.Series):
        dic = {f'group{str(g_value)}':factor_daily[factor_daily == g_value].index.tolist() for g_value in range(1, 1+group_number)}
        return pd.Series(dic)
    
    return {'group_stock':fg.apply(_factor_group, axis=1), 'group_index':fg}
```
